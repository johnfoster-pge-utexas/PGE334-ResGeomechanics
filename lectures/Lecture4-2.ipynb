## Effective stress tensor

${}$

$$
\boldsymbol{\sigma}_{eff} = 
\begin{bmatrix}
S_{11} & S_{12} & S_{13} \\
S_{12} & S_{22} & S_{23} \\
S_{13} & S_{23} & S_{33}
\end{bmatrix} -
\begin{bmatrix}
P_p & 0 & 0 \\
0 & P_p & 0 \\
0 & 0 & P_p
\end{bmatrix}
$$



${}$

$$
\boldsymbol{\sigma}_{eff}= 
\begin{bmatrix}
S_{11} - P_p & S_{12} & S_{13} \\
S_{12} & S_{22} - P_p & S_{23} \\
S_{13} & S_{23} & S_{33} - P_p
\end{bmatrix}
$$

 * Faulting depends on the effective stress

# Pore pressure at depth

${}$

$P_p^{\mbox{hydro}} = \int_0^z \rho_w(z) g {\rm d}z \approx \rho_w g z_w$

## Ratio of pore pressure to $S_v$

$\lambda_p = P_p / S_v$

**Hydrostatic:** $\lambda_p \approx 0.44$

**Lithostatic:** $\lambda_p = 1$

# Mechanisms of overpressure

# Disequilibrium compaction

 * Ongoing sedimentation increases overburden (vertical stress) faster than fluid diffuses out of zone

## Characteristic time of diffusion in porous medium

$${}$$

$$
\tau = \frac{(\phi \beta_f + \beta_r) \eta l^2}{k}
$$

 * low-permiability sand (~1 md) 
   - $\tau$ on the order of years for $l=0.1$km

 * low-permiability shale (~10 nd) 
   - $\tau$ on the order of 100,000 years for $l=0.1$km
