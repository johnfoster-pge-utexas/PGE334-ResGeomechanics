$S_{33} = S_v$ must be a principle stress!

## Four parameters needed to describe state-of-stress in the earth

 * $S_\mbox{v}$ - vertical stress magnitude

 * $S_{\mbox{Hmax}}$ - maximum horizontal principle stress magnitude

 * $S_\mbox{hmin}$ - minimum horizontal principle stress magnitude

 * One horizontal principle direction, usually the direction associated with $S_\mbox{Hmax}$

## Summary


| Regime | $S_1$ | $S_2$ | $S_3$ |
|--------|-------|------|-------|
| Normal | $S_v$ | $S_{Hmax}$ | $S_{hmin}$ |
| Strike-slip | $S_{Hmax}$ | $S_{v}$ | $S_{hmin}$ |
| Reverse | $S_{Hmax}$ | $S_{hmin}$ | $S_{v}$ |




#Vertical stress magnitude

$$
S_v =  \int_0^{z} \rho(z) g {\rm d}z 
$$

## In offshore areas

$$
S_v = \rho_w g z_w + \int_{z_w}^{z} \rho(z) g {\rm d}z
$$

## Rules of thumb

 * $\rho_w \approx 1 \mbox{g}/\mbox{cm}^3$
   - increases at a rate of 10 MPa/km (0.44 psi/ft)

 * $\rho_{\mbox{rock}} \approx 2.3 \mbox{g}/\mbox{cm}^3$
    - increaes at a rate of 23 MPa/km (1 psi/ft)
